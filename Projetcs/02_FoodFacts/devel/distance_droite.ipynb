```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
a = 0.5
b = 5

x0 = np.linspace( 0, 4, 2)

xi, yi = 3, 6

ap = -1/a

# y = a*x + b
# b = y - a*x

bp = yi + xi/a

# xp / (-a) + bp = a*xp + b
# bp - b = xp * (a + 1./ a)
# xp = ( yi + xi/a - b ) / (a + 1./a )
xp = ( yi + xi/a - b ) / (a + 1./a )
yp = a*xp + b

print( 'diff yp:', yp - (xp*ap+bp) )

print('ref:', (xi-xp)**2 + (yi-yp)**2 )
print( ( xi - xp )**2 + ( a*xp + b -yi )**2 )

fig, ax = plt.subplots()
ax.plot( x0, a*x0 + b, 'r' )
ax.plot( xi, yi, 'ro' )

ax.plot( [xi-1, xi+1], [yi+1./a, yi-1/a] , 'b' )

ax.plot( x0, x0*ap + bp , 'g--')

ax.plot( [xp], [a*xp + b], 'bo')

ax.axis('equal')
```
